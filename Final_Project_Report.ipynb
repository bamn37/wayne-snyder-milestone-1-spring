### 4.1 Analytical Framework

Describe your overall approach.

Include:
- your overall framework
- use of validation curves
- choice of MAE or RMSE (as appropriate) as the primary error metric

**4.1 Answer:**  
Our overall framework followed a fairly standard supervised machine learning pipeline for a regression problem.  First we cleaned the raw Zillow dataset, explored it through EDA, created a preprocessed modeling dataset, and then compared several candidate regression models.  From there, we introduced feature engineering, applied feature selection, tuned the most promising models, and then finalized the best overall model using both cross validation and a hold out test set to ensure no data leakage between training and testing.  The project was designed so that each stage built on the previous one rather than trying to jump directly to a final model.

For our initial model comparison we tested Linear Regression, Decision Tree Regressor, and Gradient Boosting Regressor using repeated cross validation.  This gave us a more stable estimate of generalization performance than relying on a single split alone. Later in the workflow, we kept the same general structure but refined the pipelines using engineered features and reduced feature subsets.  Hyperparameter tuning was then performed separately for each model type.  Our primary error metric was MAE (Mean Absolute Error) rather than RMSE.  The main reason is interpretability as MAE is expressed directly in the original unit of dollars.  This makes it easier to explain the model’s performance in practical terms. If a model has an MAE of about $168,000, that means its predictions are off by about that much on average. That is a more intuitive message for business users than squaredberror based metrics.  MAE is also less sensitive than RMSE to a small number of very expensive homes, which is useful here because the target distribution is strongly right skewed and includes luxury properties that can otherwise dominate the loss metric used.

Though we didn't explicitly make validation curves (plotting training error vs validation error), they were used conceptually as part of the tuning mindset even though the main implementation relied on repeated cross validation and hyperparameter search rather than a formal single validation curve plot for every model.  Basically we did the same type of thinking by using cross validation and hyperparameter tuning to essentially sample points along the validation curve.The important idea was the same: evaluate how changes in model complexity affected generalization, and avoid choosing a model that only looked good on training data.  This was especially relevant for the Decision Tree and Gradient Boosting models, where depth, leaf size, and ensemble settings directly control overfitting behavior.

### 4.2 Data Cleaning and Preprocessing

Describe how you prepared the data.

Include:
- issues in the raw data
- handling of missing values, outliers, inconsistencies
- key decisions and why
- what worked and what did not work

**4.2 Answer:**  
The raw data required a great deal of cleaning before modeling.  One of the first problems was that several features were not useful for the business objective even before considering missingness.  For example, parcelid is just a unique identifier and has no predictive meaning for home value.  assessmentyear is effectively constant, and some geographic text/code fields such as propertycountylandusecode and propertyzoningdesc were high cardinality fields that would add a lot of complexity without clear benefit.  Those conceptually weak columns were dropped first.

A second major issue was missing data.  Many features were missing for a very large percentage of rows. We used a threshold of 60% missing to drop features that were too sparse to be worth imputing.  The reasoning was that if most of a column is empty, then filling it in would mean inventing more data than preserving, which would probably hurt more than help.  This step reduced the dataset from 49 columns to 20 columns after the earlier conceptual drops.

We also removed problematic samples at the row level.  Rows with a missing target value were dropped immediately because they are unusable in supervised learning.  In addition, we removed extreme outliers in the target by dropping the bottom 0.1% and top 0.5% of taxvaluedollarcnt(see boxplot below). This was done because very small values may represent data quality problems, while the most extreme high end properties behave differently from typical homes and can distort model fitting, especially for error based objective functions.

For the remaining missing values, we imputed using the median for numeric columns.  This choice made sense because many of the key variables, such as size and count related fields, were right skewed and the median is more robust than the mean.  This worked well and produced a fully numeric cleaned dataset with no null values remaining.  One thing that did not turn out to be necessary was extensive categorical encoding, because after cleaning, the remaining variables were already numeric or numeric coded.  Overall, the preprocessing choices seem to have helped rather than hurt because they reduced noise, preserved most of the usable data, and produced a stable dataset for modeling.



### 4.3 Feature Engineering

Describe your feature engineering.

Include:
- transformations or new features
- why you created them
- which were useful or not useful
- what worked and what did not work

**4.3 Answer:**  
Feature engineering focused on creating transformations that might better expose useful relationships between housing characteristics and property value.  Based on the EDA from Milestone 1, several important predictors were strongly right skewed, especially size related variables.  Because of that, we engineered log_sqft from calculatedfinishedsquarefeet and log_lotsize from lotsizesquarefeet.  The idea was to reduce skew and make these predictors more linear or at least more stable for the models.  We also created bath_bed_ratio, defined as bathroom count divided by bedroom count plus one. The reasoning there was that the relationship between bathrooms and bedrooms may provide more information about a home’s quality or livability than either count alone.

We also explored another engineered feature, house age, but it did not appear to add as much signal as expected. In contrast, log_sqft was consistently useful throughout the project.  It was retained in later feature selection and appeared in both the tree based and linear model subsets, which is a good sign that it captured real signal rather than noise. bath_bed_ratio also showed some usefulness, especially for the tree based models, which suggests it may be capturing an interaction or non linear relationship that simple raw counts do not express as clearly. log_lotsize was  less helpful. It contributed some value, but not as strongly as log_sqft.

In terms of measured impact, the overall gains from feature engineering were modest rather than dramatic. That is not surprising since the baseline models were already working with fairly informative housing features, so feature engineering was more about small improvements and better representation than about changing the problem entirely.  The biggest improvement was seen for Linear Regression, which improved by about $1,300 in mean CV MAE after adding engineered features. This makes sense because linear models benefit more directly from transformations that reduce skew and improve linearity. Tree based models improved very little, which also makes sense because they are generally less sensitive to monotonic transformations.

What worked best was adding a small number of targeted, interpretable transformations rather than creating many synthetic features.  The project ended up confirming that some transformations help a lot more than others, and that usefulness depends on the model family being used.

### 4.4 Model Selection

Describe how you selected your model.

Include:
- models you tried and why
- how you evaluated generalization
- hyperparameter tuning
- how you chose the final model

**4.4 Answer:**  
We began model selection by comparing three regression models that represent different levels of flexibility:  Linear Regression, Decision Tree Regressor, and Gradient Boosting Regressor.  Linear Regression provided a simple interpretable baseline, the Decision Tree allowed for non linear splits and interactions, and Gradient Boosting represented a stronger ensemble approach that often performs well on tabular structured data (this type of model often dominates the leaderboard of kaggle competitions).  We evaluated all three using repeated cross validation on the training data so that comparisons would reflect generalization rather than a lucky or unlucky single split.

The initial baseline results already made the ranking fairly clear. Gradient Boosting performed best with a mean CV MAE of about $179k, followed by Linear Regression at about $206k, while the Decision Tree performed worst at about $231k.  This pattern was consistent with expectations.  A single tree can easily overfit and tends to be less stable, while boosting usually captures more complex structure without giving up too much bias control.

After we carried all three models through the same general improvement process.  We added engineered features, then applied feature selection. For feature selection, we used SelectKBest with F-regression for the linear model and tree-based feature importance for the tree models, keeping the top 15 features in each case.  The resulting subsets were similar across model types, with about 11 shared features.  That gave us some confidence that the strongest predictors were consistent regardless of modeling style.  Even after these adjustments, Gradient Boosting remained the best model, and the feature engineering and selection stages mostly produced marginal refinements rather than changing the overall ranking.

For hyperparameter tuning, we used a model specific strategy.  We replaced plain Linear Regression with Ridge Regression and tuned the regularization strength using GridSearchCV. For the Decision Tree and Gradient Boosting models, we used RandomizedSearchCV because their hyperparameter spaces were larger and more computationally expensive to search exhaustively.  Important parameters included tree depth, minimum samples per split, minimum samples per leaf, pruning for the tree model, learning rate, number of estimators, depth, and subsampling for Gradient Boosting.

The final model was selected based on repeated cross validated MAE, stability across stages, and holdout test performance.  Gradient Boosting was chosen because it consistently had the lowest validation error and also generalized well on the test set, giving the strongest overall balance between flexibility and predictive accuracy.
